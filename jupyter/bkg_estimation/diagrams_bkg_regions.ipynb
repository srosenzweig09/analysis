```python
import matplotlib.patches as mpatches
```

## Concentric

```python
h_signal = 'oo'
h_control = 'XX'

signal_region = mpatches.Patch(color='black', fill=False, hatch=h_signal, alpha=0.5, label='Signal Region')
control_region = mpatches.Patch(color='black', fill=False, hatch=h_control, alpha=0.5, label='Control Region')
```

```python
SR_color = 'goldenrod'
CR_color = 'steelblue'
SR_text = 'black'
CR_text = CR_color

Ac = 125

fig, ax = plt.subplots()

circle3 = plt.Circle((Ac, Ac), 25, alpha=0.5, color=SR_color)
circle_a = plt.Circle((Ac, Ac), 25, color='white')
circle4 = plt.Circle((Ac, Ac), 50, alpha=0.5, color=CR_color)

ax.add_patch(circle4)
ax.add_patch(circle_a)
ax.add_patch(circle3)

ax.text(Ac, Ac, 'ASR', ha='center', va='center', fontsize=16, color='black')
ax.text(152, 152, 'ACR', ha='right', va='bottom', fontsize=16, color='black')

ax.set_ylim(0, 250)
ax.set_xlim(0, 250)

ax.set_xlabel(r"M$_\mathrm{H,i}$ [GeV]", fontsize=26)
ax.set_ylabel(r"M$_\mathrm{H,j}$ [GeV]", fontsize=26)

ax.set_title("Analysis Region")

sr_patch = mpatches.Patch(color=SR_color, label='Signal Region', alpha=0.5)
cr_patch = mpatches.Patch(color=CR_color, label='Control Region', alpha=0.5)
plt.legend(handles=[sr_patch, cr_patch])

plt.xticks(list(plt.xticks()[0]) + [125])
plt.yticks(list(plt.yticks()[0]) + [125])

fig.savefig("plots/diagrams/bkg_model/concentric/analysis_region.pdf")
```

```python
SR_color = 'peru'
CR_color = 'cadetblue'
SR_text = 'black'
CR_text = CR_color

Ac = 125

fig, ax = plt.subplots()

circle3 = plt.Circle((Ac, Ac), 50, alpha=0.5, color=SR_color)
circle_a = plt.Circle((Ac, Ac), 25, color='white')
circle4 = plt.Circle((Ac, Ac), 75, alpha=0.5, color=CR_color)
circle_b = plt.Circle((Ac, Ac), 50, color='white')

ax.add_patch(circle4)
ax.add_patch(circle_b)
ax.add_patch(circle3)
ax.add_patch(circle_a)

ax.text(100, 150, 'VSR', ha='center', va='center', fontsize=16, color='black')
ax.text(150, 177, 'VCR', ha='right', va='bottom', fontsize=16, color='black')

ax.set_ylim(0, 250)
ax.set_xlim(0, 250)

ax.set_xlabel(r"M$_\mathrm{H,i}$ [GeV]", fontsize=26)
ax.set_ylabel(r"M$_\mathrm{H,j}$ [GeV]", fontsize=26)

ax.set_title("Validation Region")

sr_patch = mpatches.Patch(color=SR_color, label='"Signal" Region', alpha=0.5)
cr_patch = mpatches.Patch(color=CR_color, label='"Control" Region', alpha=0.5)
plt.legend(handles=[sr_patch, cr_patch])

plt.xticks(list(plt.xticks()[0]) + [125])
plt.yticks(list(plt.yticks()[0]) + [125])

fig.savefig("plots/diagrams/bkg_model/concentric/validation_region.pdf")
```

```python
SR_color = 'green'
CR_color = 'orange'
VCR_color = 'mediumpurple'

Ac = 125

fig, ax = plt.subplots()

asr = plt.Circle((Ac, Ac), 25, color=SR_color, alpha=0.7)
asr_block = plt.Circle((Ac, Ac), 25, color='white')
acr = plt.Circle((Ac, Ac), 50, color=CR_color, alpha=0.8)
acr_block = plt.Circle((Ac, Ac), 50, color='white')
vcr = plt.Circle((Ac, Ac), 75, color=VCR_color)

for c in [vcr, acr_block, acr, asr_block, asr]:
    ax.add_patch(c)

ax.text(Ac, Ac, 'ASR', ha='center', va='center', fontsize=22, color='black')
ax.text(125, 160, 'ACR / VSR', ha='center', va='center', fontsize=22, color='black')
ax.text(125, 185, 'VCR', ha='center', va='center', fontsize=22, color='black')

ax.set_ylim(25, 225)
ax.set_xlim(25, 225)

ax.set_xlabel(r"M$_\mathrm{H,i}$ [GeV]", fontsize=26)
ax.set_ylabel(r"M$_\mathrm{H,j}$ [GeV]", fontsize=26)

ax.set_xticks([125])
ax.set_yticks([125])

ax.plot([125,125],[25,50], color='lightgray', linestyle='--')
ax.plot([25,50],[125,125], color='lightgray', linestyle='--')

ax.set_title("Analysis and Validation Regions")
fig.savefig("plots/diagrams/bkg_model/concentric/all_regions.pdf")
```

```python
SR_color = 'green'
CR_color = 'orange'
VCR_color = 'mediumpurple'
Ac = 125

fig, axs = plt.subplots(ncols=3, figsize=(18,6))

ax = axs[0]
asr = plt.Circle((Ac, Ac), 25, color=SR_color, alpha=0.7)
asr_block = plt.Circle((Ac, Ac), 25, color='white')
acr = plt.Circle((Ac, Ac), 60, color=CR_color, alpha=0.8)
acr_block = plt.Circle((Ac, Ac), 50, color='white')
vcr = plt.Circle((Ac, Ac), 75, color=VCR_color)

ax.set_title("Increase CR outer edge", fontsize=18)
for c in [vcr, acr_block, acr, asr_block, asr]:
    ax.add_patch(c)

ax = axs[1]
asr = plt.Circle((Ac, Ac), 25, color=SR_color, alpha=0.7)
asr_block = plt.Circle((Ac, Ac), 25, color='white')
acr = plt.Circle((Ac, Ac), 40, color=CR_color, alpha=0.8)
acr_block = plt.Circle((Ac, Ac), 50, color='white')
vcr = plt.Circle((Ac, Ac), 75, color=VCR_color)

ax.set_title("Decrease CR outer edge", fontsize=18)
for c in [vcr, acr_block, acr, asr_block, asr]:
    ax.add_patch(c)

ax = axs[2]
asr = plt.Circle((Ac, Ac), 25, color=SR_color, alpha=0.7)
asr_block = plt.Circle((Ac, Ac), 30, color='white')
acr = plt.Circle((Ac, Ac), 50, color=CR_color, alpha=0.8)
acr_block = plt.Circle((Ac, Ac), 50, color='white')
vcr = plt.Circle((Ac, Ac), 75, color=VCR_color)

ax.set_title("Decrease CR inner edge", fontsize=18)
for c in [vcr, acr_block, acr, asr_block, asr]:
    ax.add_patch(c)

for ax in axs.flatten():
    ax.text(Ac, Ac, 'ASR', ha='center', va='center', fontsize=16, color='black')
    ax.text(125, 160, 'ACR / VSR', ha='center', va='center', fontsize=16, color='black')
    ax.text(125, 185, 'VCR', ha='center', va='center', fontsize=16, color='black')

    ax.set_ylim(25, 225)
    ax.set_xlim(25, 225)

    ax.set_xlabel(r"M$_\mathrm{H,i}$ [GeV]", fontsize=18)
    ax.set_ylabel(r"M$_\mathrm{H,j}$ [GeV]", fontsize=18)

    ax.set_xticks([125])
    ax.set_yticks([125])

    ax.plot([125,125],[25,50], color='lightgray', linestyle='--')
    ax.plot([25,50],[125,125], color='lightgray', linestyle='--')

plt.tight_layout()
# ax.set_title("Analysis and Validation Regions")
# fig.savefig("plots/diagrams/bkg_model/concentric/all_regions.pdf")
```
