```python
def get_mx(signal):
    mx = int(signal.split('/')[-2].split('_')[2].split('-')[1])
    # my = int(signal.split('/')[-2].split('_')[3].split('-')[1])
    return mx
```

```python
mx = []
my = []
eff = []

for signal in gnn_signal:
    # eff.append(ak.sum(signal.resolved_mask)/len(signal.resolved_mask))
    eff.append(ak.sum(signal.resolved_mask)/signal.cutflow[0])
    mx.append(signal.mx)
    my.append(signal.my)

data = np.column_stack((mx, my, eff))

df = DataFrame(data, columns=['MX', 'MY', 'Efficiency'])
# df = df.convert_dtypes()
df = df.astype({'MX':'int'})
df = df.astype({'MY':'int'})
data = df.pivot(index='MY', columns='MX', values='Efficiency')
data = data.sort_values(by=['MY'], ascending=False)
```

```python
fname = '/uscms/home/srosenzw/nobackup/workarea/higgs/sixb_analysis/CMSSW_10_6_19_patch2/src/sixb/eightbStudies/scripts/ntuple_filelist.txt'
with open(fname) as f:
    files = f.readlines()
files = [f.rstrip('\n') for f in files if 'NMSSM' in f]
```

```python
training = [SixB(f, gnn_model=False) for f in files]
nevents = [[t.nevents, ak.sum(t.resolved_mask)] for t in training]
nevents = np.asarray(nevents)
```

```python
MX, MY, total, resolved = [], [], [], []
for n1,sample in zip(nevents, training):
    print(sample.mx, sample.my, n1)
    MX.append(sample.mx)
    MY.append(sample.my)
    total.append(n1[0])
    resolved.append(n1[1])
```

```python
len(training)
```

```python
data = np.column_stack((MX, MY, resolved, total))

df = DataFrame(data, columns=['MX', 'MY', 'Training', 'Total'])
# df = df.convert_dtypes()
df = df.astype({'MX':'int'})
df = df.astype({'MY':'int'})
training_data = df.pivot(index='MY', columns='MX', values='Training')
training_data = training_data.sort_values(by=['MY'], ascending=False)
produced_data = df.pivot(index='MY', columns='MX', values='Total')
produced_data = produced_data.sort_values(by=['MY'], ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(20,17))

ax = heatmap(training_data, annot=True, ax=ax, annot_kws={"fontsize":14})
ax.set_xlabel(r"$M_X$ [GeV]")
ax.set_ylabel(r"$M_Y$ [GeV]")

ax.set_title('Number of Training Events')

fig.savefig(f'plots/1_signal_exploration/resolved/training_events')
```
