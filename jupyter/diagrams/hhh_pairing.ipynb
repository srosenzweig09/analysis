```python
from matplotlib.patches import Circle
```

```python
def f(x):
    return 0.77778*x
```

```python
fig, ax = plt.subplots()

x_a, y_a = 50, 75
x_b, y_b = 30, 45
x_c, y_c = 70, 35

color_a = 'firebrick'
color_b = 'goldenrod'
color_c = 'forestgreen'

ax.plot([0,90],[0,70],'--',color='k')
ax.set_axis_off()
ax.arrow(0,0,100,0,head_width=1,color='k')
ax.arrow(0,0,0,100,head_width=1,color='k')

ax.text(-10, 103, r'$M(H_X)$ [GeV]')
ax.text(80, -5, r'$M(H_1)$ [GeV]')

x = 67
ax.plot([x_a,x],[y_a,f(x)],'--',color=color_a)
x = 40
ax.plot([x_b,x],[y_b,f(x)],'--',color=color_b)
x = 61
ax.plot([x_c,x],[y_c,f(x)],'--',color=color_c)

circle_a = Circle((x_a,y_a,5), facecolor=color_a, edgecolor='k', zorder=2)
circle_b = Circle((x_b,y_b,5), facecolor=color_b, edgecolor='k', zorder=2)
circle_c = Circle((x_c,y_c,5), facecolor=color_c, edgecolor='k', zorder=2)

ax.add_artist(circle_a)
ax.add_artist(circle_b)
ax.add_artist(circle_c)

ax.text(x_a,y_a,'A',ha='center',va='center',zorder=3)
ax.text(x_b,y_b,'B',ha='center',va='center',zorder=3)
ax.text(x_c,y_c,'C',ha='center',va='center',zorder=3)

ax.text(0,0,'(0,0)',ha='right',va='top')

ax.text(x_a + 10, y_a - 11, r"$D_{HHH}^3$",color=color_a)
ax.text(x_b - 5, y_b - 12, r"$D_{HHH}^2$",color=color_b)
ax.text(x_c - 15, y_c + 3, r"$D_{HHH}^1$",color=color_c)

fig.savefig("plots/diagrams/dhhh_pairing.pdf")
```
