### Bkg Estimation-Related Distributions

```python
# def get_region_masks(HX_m, H1_m, H2_m):
def get_region_masks(tree):
    dM_125 = np.sqrt((tree.HX.m - 125)**2 + (tree.H1.m - 125)**2 + (tree.H2.m - 125)**2)
    asr_mask = dM_125 < 25
    acr_mask = ~asr_mask & (dM_125 < 50)
    return asr_mask, acr_mask
```

```python
gnn_asr_mask, gnn_acr_mask = get_region_masks(gnn_bkg)
bias_asr_mask, gnn_acr_mask = get_region_masks(bias_bkg)

gnn_dm_125 = np.sqrt((gnn_bkg.HX.m - 125)**2 + (gnn_bkg.H1.m - 125)**2 + (gnn_bkg.H2.m - 125)**2)
gnn_asr_mask = gnn_dm_125 < 30
gnn_acr_mask = ~gnn_asr_mask & (gnn_dm_125 < 50)

bias_dm_125 = np.sqrt((bias_bkg.HX.m - 125)**2 + (bias_bkg.H1.m - 125)**2 + (bias_bkg.H2.m - 125)**2)
bias_asr_mask = bias_dm_125 < 30
bias_acr_mask = ~bias_asr_mask & (bias_dm_125 < 50)

gnn_dm2d_125 = np.sqrt((gnn_bkg.H1.m - 125)**2 + (gnn_bkg.H2.m - 125)**2)
gnn_2d_asr_mask = gnn_dm2d_125 < 30
gnn_2d_acr_mask = ~gnn_2d_asr_mask & (gnn_dm2d_125 < 50)

gnn_dm1d_125 = np.sqrt((gnn_bkg.HX.m - 125)**2)
gnn_1d_asr_mask = gnn_dm1d_125 < 30
gnn_1d_acr_mask = ~gnn_1d_asr_mask & (gnn_dm1d_125 < 50)
```
