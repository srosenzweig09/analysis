## 2. Define control regions

### <span style="font-variant:small-caps;"><font size=4>Control Region</font></span>: <font size=3>Thresholds</font>

```python
cut_6jNN = 0.8
mass_veto = 60
cut_sum_btag = 0.64
```

### <span style="font-variant:small-caps;"><font size=4>Control Region</font></span>: <font size=3>Masks</font>

```python
sgnl_6jNN_mask = signal_tree.b_6j_score > cut_6jNN
sgnl_pass_mass_mask = abs(signal_tree.t6_higgs_m[:,2] - 125) > mass_veto
sgnl_fail_btag_mask = ak.sum(signal_tree.t6_jet_btag, axis=1)/6 < cut_sum_btag

data_6jNN_mask = data_tree.b_6j_score > cut_6jNN
data_fail_mass_mask = abs(data_tree.t6_higgs_m[:,2] - 125) > mass_veto
data_fail_btag_mask = ak.sum(data_tree.t6_jet_btag, axis=1)/6 < cut_sum_btag
```

```python
sgnl_pass_mass = sgnl_6jNN_mask & ~sgnl_pass_mass_mask
sgnl_pass_mass = sgnl_6jNN_mask & sgnl_pass_mass_mask
data_fail_mass = data_6jNN_mask & data_fail_mass_mask
```

```python
sgnl_pass_mass_btag = sgnl_pass_mass & ~sgnl_fail_btag_mask
sgnl_pass_mass_btag = sgnl_pass_mass & sgnl_fail_btag_mask

data_fail_mass_fail_btag = data_6jNN_mask & data_fail_mass_mask & data_fail_btag_mask
data_fail_mass_pass_btag = data_6jNN_mask & data_fail_mass_mask & ~data_fail_btag_mask
data_pass_mass_fail_btag = data_6jNN_mask & ~data_fail_mass_mask & data_fail_btag_mask
```
