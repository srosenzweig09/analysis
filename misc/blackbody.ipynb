```python
from scipy.constants import c, h, k, e

def radiance(nu, T):
    coeff = 2 * h * nu**3 / c**2
    denom = h * nu / (e * k * T) - 1
    return coeff / denom

def planck(wavelength, T):
    # wavelength in um
    wavelength = wavelength * 1e-6
    coeff = 8 * np.pi * h * c / wavelength**5
    denom = np.exp(h * c / (k * T * wavelength))
    return coeff / denom
```

```python
fig, ax = plt.subplots()

x = np.linspace(0, 40, 1000)
y_300 = planck(x, 300)
y_400 = planck(x, 400)
y_500 = planck(x, 500)
y_600 = planck(x, 600)

ax.plot(x, y_300, label='T = 300 K')
ax.plot(x, y_400, label='T = 400 K')
ax.plot(x, y_500, label='T = 500 K')
ax.plot(x, y_600, label='T = 600 K')

ax.set_xlabel(r'Wavelength $\lambda$ [$\micro$,]')

ax.legend()
```
